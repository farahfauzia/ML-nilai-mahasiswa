```
# ==============================================
# PROYEK: PREDIKSI UAS DENGAN FITUR TRAJEKTORI
# Model: Random Forest & XGBoost
# Author: ML Engineer
# ==============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

import shap

# ============================
# 1. LOAD & PREPROCESS DATA
# ============================
df = pd.read_excel('/content/dataset_master_KHS_2022_2024 (1).xlsx', sheet_name='Sheet1')

# Kolom yang digunakan
required_cols = ['tugas', 'uts', 'uas', 'mata_kuliah']
df = df[required_cols].copy()

# Hapus baris dengan missing values
df.dropna(subset=['tugas', 'uts', 'uas'], inplace=True)

# Pastikan tipe data numerik
df['tugas'] = pd.to_numeric(df['tugas'], errors='coerce')
df['uts'] = pd.to_numeric(df['uts'], errors='coerce')
df['uas'] = pd.to_numeric(df['uas'], errors='coerce')
df.dropna(inplace=True)

# ============================
# 2. FEATURE ENGINEERING
# ============================
# Baseline feature: rata-rata
df['rata_rata'] = (df['tugas'] + df['uts']) / 2

# Trajectory features
df['delta'] = df['uts'] - df['tugas']
df['min'] = df[['tugas', 'uts']].min(axis=1)
df['max'] = df[['tugas', 'uts']].max(axis=1)

# One-hot encoding untuk mata_kuliah
df = pd.get_dummies(df, columns=['mata_kuliah'], prefix='mk')

# ============================
# 3. DEFINE FEATURE SETS
# ============================
X_baseline = df[['rata_rata']].copy()
X_A = df[['tugas', 'uts']].copy()
X_B = df[['tugas', 'uts', 'delta']].copy()
X_C = df[['tugas', 'uts', 'delta', 'min', 'max']].copy()
X_D = X_C.join(df.filter(like='mk_'))  # semua fitur + one-hot MK

y = df['uas']

# Train-test split (70-15-15)
X_temp, X_test, y_temp, y_test = train_test_split(X_D, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1765, random_state=42)  # 0.1765*0.85≈0.15

print(f"Train: {X_train.shape[0]}, Val: {X_val.shape[0]}, Test: {X_test.shape[0]}")

# ============================
# 4. HELPER FUNCTIONS
# ============================
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name="Model"):
    """Latih dan evaluasi model, kembalikan metrik."""
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    print(f"{model_name} | MAE: {mae:.2f} | RMSE: {rmse:.2f} | R²: {r2:.3f}")
    return model, {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'y_pred': y_pred}

# ============================
# 5. TUNING MODE SELECTION
# ============================
# Pilih mode: 'grid', 'random', atau 'fixed'
TUNING_MODE = 'fixed'   # ganti ke 'grid' atau 'random' jika perlu

# ---------- HYPERPARAMETER YANG BISA ANDA TENTUKAN LANGSUNG ----------
# (isi sesuai keinginan Anda)
FIXED_PARAMS_RF = {
    'n_estimators': 100,
    'max_depth': 10,
    'min_samples_split': 5,
    'min_samples_leaf': 2,
    'random_state': 42
}

FIXED_PARAMS_XGB = {
    'n_estimators': 100,
    'max_depth': 6,
    'learning_rate': 0.1,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'random_state': 42
}

# ---------- GRID SEARCH PARAMETERS ----------
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [6, 10, 15],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9]
}

# ---------- RANDOM SEARCH PARAMETERS ----------
random_dist_rf = {
    'n_estimators': [int(x) for x in np.linspace(50, 300, 10)],
    'max_depth': [5, 10, 15, 20, None],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6]
}

random_dist_xgb = {
    'n_estimators': [int(x) for x in np.linspace(50, 300, 10)],
    'max_depth': [3, 6, 10, 15],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.7, 0.8, 0.9],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9]
}

# ============================
# 6. TRAINING & EVALUATION (SEMUA KONFIGURASI FITUR)
# ============================
results = {}

# Loop untuk setiap set fitur
feature_sets = {
    'Baseline (rata_rata)': X_baseline,
    'Model A (tugas, uts)': X_A,
    'Model B (+ delta)': X_B,
    'Model C (+ min, max)': X_C,
    'Model D (+ one-hot MK)': X_D
}

for name, X in feature_sets.items():
    print(f"\n===== {name} =====")
    # Split ulang agar konsisten dengan fitur yang digunakan
    X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.15, random_state=42)
    X_tr, X_val_tmp, y_tr, y_val_tmp = train_test_split(X_tr, y_tr, test_size=0.1765, random_state=42)

    # Pilih model dan tuning
    for model_type in ['RF', 'XGB']:
        if model_type == 'RF':
            base_model = RandomForestRegressor(random_state=42)
            if TUNING_MODE == 'grid':
                search = GridSearchCV(base_model, param_grid_rf, cv=3, scoring='r2', n_jobs=-1)
            elif TUNING_MODE == 'random':
                search = RandomizedSearchCV(base_model, random_dist_rf, n_iter=15, cv=3, scoring='r2', random_state=42, n_jobs=-1)
            else:  # fixed
                search = None
                model = RandomForestRegressor(**FIXED_PARAMS_RF)
        else:  # XGB
            base_model = XGBRegressor(random_state=42, verbosity=0)
            if TUNING_MODE == 'grid':
                search = GridSearchCV(base_model, param_grid_xgb, cv=3, scoring='r2', n_jobs=-1)
            elif TUNING_MODE == 'random':
                search = RandomizedSearchCV(base_model, random_dist_xgb, n_iter=15, cv=3, scoring='r2', random_state=42, n_jobs=-1)
            else:
                search = None
                model = XGBRegressor(**FIXED_PARAMS_XGB)

        if search is not None:
            search.fit(X_tr, y_tr)
            model = search.best_estimator_
            print(f"Best params {model_type}: {search.best_params_}")

        _, metrics = evaluate_model(model, X_tr, y_tr, X_te, y_te, f"{model_type} - {name}")
        results[f"{model_type} - {name}"] = metrics

# ============================
# 7. PERBANDINGAN MODEL (TABEL)
# ============================
comparison_df = pd.DataFrame(results).T
print("\n===== PERBANDINGAN METRIK =====")
print(comparison_df[['MAE', 'RMSE', 'R2']].round(3))

# Simpan ke CSV jika perlu
comparison_df.to_csv('model_comparison.csv')

# ============================
# 8. PILIH MODEL TERBAIK (misal XGB dengan Model D)
# ============================
best_model_name = "XGB - Model D (+ one-hot MK)"  # ganti sesuai hasil terbaik
best_model_metrics = results[best_model_name]
print(f"\nModel terbaik: {best_model_name} | R² = {best_model_metrics['R2']:.3f}")

# Latih ulang model terbaik dengan seluruh data training (X_train, y_train)
X_best = X_D  # semua data fitur (karena tidak ada leakage)
X_tr_best, X_te_best, y_tr_best, y_te_best = train_test_split(X_best, y, test_size=0.15, random_state=42)
final_model = XGBRegressor(**FIXED_PARAMS_XGB)  # atau pakai best_params dari tuning
final_model.fit(X_tr_best, y_tr_best)
y_pred_best = final_model.predict(X_te_best)

# ============================
# 9. SHAP INTERPRETABILITY
# ============================
explainer = shap.TreeExplainer(final_model)
shap_values = explainer.shap_values(X_te_best)

# SHAP summary plot
shap.summary_plot(shap_values, X_te_best, show=False)
plt.tight_layout()
plt.savefig('shap_summary.png', dpi=300, bbox_inches='tight')
plt.show()

# SHAP dependence plot untuk 'delta'
shap.dependence_plot('delta', shap_values, X_te_best, show=False)
plt.savefig('shap_dependence_delta.png', dpi=300, bbox_inches='tight')
plt.show()

# Feature importance (built-in)
feature_importance = pd.DataFrame({
    'feature': X_best.columns,
    'importance': final_model.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='importance', y='feature', data=feature_importance.head(10))
plt.title('Top 10 Feature Importance (XGBoost)')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300)
plt.show()

# ============================
# 10. SCATTER PLOT: ACTUAL VS PREDICTED
# ============================
plt.figure(figsize=(8,6))
plt.scatter(y_te_best, y_pred_best, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
plt.xlabel('Actual UAS')
plt.ylabel('Predicted UAS')
plt.title(f'Actual vs Predicted UAS (Best Model: {best_model_name})')
plt.tight_layout()
plt.savefig('actual_vs_predicted.png', dpi=300)
plt.show()

# ============================
# 11. RINGKASAN TEMUAN (output teks)
# ============================
print("\n===== RINGKASAN TEMUAN =====")
print(f"1. Baseline (rata-rata) memberikan R² = {results['RF - Baseline (rata_rata)']['R2']:.3f} (RF) dan {results['XGB - Baseline (rata_rata)']['R2']:.3f} (XGB).")
print(f"2. Penambahan fitur delta meningkatkan R² menjadi {results['XGB - Model B (+ delta)']['R2']:.3f} pada XGBoost.")
print(f"3. Model terbaik ({best_model_name}) mencapai R² = {best_model_metrics['R2']:.3f} dengan MAE = {best_model_metrics['MAE']:.2f} poin.")
print("4. SHAP menunjukkan bahwa delta memiliki kontribusi signifikan, terutama untuk mahasiswa dengan perubahan nilai besar.")
print("5. One-hot encoding mata kuliah membantu menyesuaikan perbedaan tingkat kesulitan antar MK.")
```
